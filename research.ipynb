```python
import pandas as pd
df = pd.read_csv("Pan-India_Bus_Routes.csv")
print(df.head())
```

```python
def extract_ac(bus_type):
    bus_type = str(bus_type).lower()

    # Check non-ac first
    if "non a/c" in bus_type or "non/ac" in bus_type or "non ac" in bus_type:
        return 0

    # Check AC
    if "a/c" in bus_type or " ac" in bus_type:
        return 1

    return 0

df["AC"] = df["Bus Type"].apply(extract_ac)
```

```python
def extract_type(bus_type):
    bus_type = str(bus_type).lower()

    if "semi sleeper" in bus_type:
        return "Semi Sleeper"
    elif "sleeper" in bus_type:
        return "Sleeper"
    elif "seater" in bus_type:
        return "Seater"
    else:
        return "Unknown"

df["Type"] = df["Bus Type"].apply(extract_type)
```

```python
import re
def extract_seats(bus_type):
    bus_type = str(bus_type)

    # Find patterns like "35 Seat"
    match = re.search(r'(\d+)\s*Seat', bus_type, re.IGNORECASE)

    if match:
        return int(match.group(1))

    # Find patterns like "33 S.S"
    match = re.search(r'(\d+)\s*S\.?S', bus_type, re.IGNORECASE)

    if match:
        return int(match.group(1))

    return None

df["No_of_Seats"] = df["Bus Type"].apply(extract_seats)
```

```python
print(df.head())
```

```python
# Fill missing seat values with average (mean)
avg_seats = df["No_of_Seats"].mean()

df["No_of_Seats"] = df["No_of_Seats"].fillna(round(avg_seats))

# Fill missing/Unknown type with mode value
mode_type = df[df["Type"] != "Unknown"]["Type"].mode()[0]

df["Type"] = df["Type"].replace("Unknown", mode_type)

# Display updated dataframe
print(df.head())
```

```python
# Function to assign pricing
def calculate_pricing(row):
    distance = row["Distance"]
    ac = row["AC"]

    # 0 - 500
    if distance <= 500:
        return 1000 if ac == 1 else 600

    # 500 - 1500
    elif distance <= 1500:
        return 2500 if ac == 1 else 1500

    # 1500 - 2500
    elif distance <= 2500:
        return 4000 if ac == 1 else 2000

    # Above 2500
    else:
        return 5500 if ac == 1 else 2500


# Create new Pricing column
df["Pricing"] = df.apply(calculate_pricing, axis=1)

# Display result
print(df.head())
```

```python
# Count frequency of each route
route_frequency = (
    df.groupby(["From", "To"])
      .size()
      .reset_index(name="Frequency")
      .sort_values(by="Frequency", ascending=False)
)

# Display most frequent routes
print(route_frequency.head())

# Most frequent route
most_frequent_route = route_frequency.iloc[0]

print("\nMost Frequent Route:")
print(f"From: {most_frequent_route['From']}")
print(f"To: {most_frequent_route['To']}")
print(f"Frequency: {most_frequent_route['Frequency']}")
```

```python
# Create Route column
df["Route"] = df["From"] + " -> " + df["To"]

# Create Frequency column
df["Frequency"] = df.groupby("Route")["Route"].transform("count")

# Keep only rows where frequency >= 10
df = df[df["Frequency"] >= 10]

# Display updated dataframe
print(df.head())

# Optional: check unique routes with frequencies
print(df[["Route", "Frequency"]].drop_duplicates())
```

```python
# Create column for number of unique bus operators per route
df["No_of_Bus_Operators"] = (
    df.groupby("Route")["Operator"]
      .transform("nunique")
)

# Display result
print(df[["Route", "Operator", "No_of_Bus_Operators"]].head())
```

```python
# Calculate average pricing per route
route_analysis = (
    df.groupby("Route")
    .agg({
        "Pricing": "mean",
        "Frequency": "first",
        "No_of_Bus_Operators": "first",
        "Distance": "mean"
    })
    .reset_index()
)

# Create a custom score
# High pricing + High frequency + Low competition

route_analysis["Opportunity_Score"] = (
    (route_analysis["Pricing"] * route_analysis["Frequency"]) /
    route_analysis["No_of_Bus_Operators"]
)

# Sort by best opportunity
top_routes = route_analysis.sort_values(
    by="Opportunity_Score",
    ascending=False
).head(10)

# Display top 10 low competition high margin routes
print(top_routes)
```

```python
# Group routes by pricing range and get top 10 routes in each range

price_ranges = {
    "0-1000": (0, 1000),
    "1000-2500": (1000, 2500),
    "2500-4000": (2500, 4000),
    "4000+": (4000, float("inf"))
}

# Create route analysis table
route_analysis = (
    df.groupby("Route")
    .agg({
        "Pricing": "mean",
        "Frequency": "first",
        "No_of_Bus_Operators": "first",
        "Distance": "mean"
    })
    .reset_index()
)

# Opportunity Score
route_analysis["Opportunity_Score"] = (
    (route_analysis["Pricing"] * route_analysis["Frequency"]) /
    route_analysis["No_of_Bus_Operators"]
)

# Find top 10 in each pricing range
for range_name, (low, high) in price_ranges.items():

    filtered = route_analysis[
        (route_analysis["Pricing"] >= low) &
        (route_analysis["Pricing"] < high)
    ]

    top_10 = filtered.sort_values(
        by="Opportunity_Score",
        ascending=False
    ).head(10)

    print(f"\n========== TOP 10 ROUTES : {range_name} ==========")
    print(top_10[[
        "Route",
        "Pricing",
        "Frequency",
        "No_of_Bus_Operators",
        "Opportunity_Score"
    ]])
```

```python
import pandas as pd

# Pricing ranges
price_ranges = {
    "0-1000": (0, 1000),
    "1000-2500": (1000, 2500),
    "2500-4000": (2500, 4000),
    "4000+": (4000, float("inf"))
}

# Route analytics
route_analysis = (
    df.groupby("Route")
    .agg({
        "Pricing": "mean",
        "Frequency": "first",
        "No_of_Bus_Operators": "first",
        "Distance": "mean"
    })
    .reset_index()
)

# Opportunity Score
route_analysis["Opportunity_Score"] = (
    (route_analysis["Pricing"] * route_analysis["Frequency"]) /
    route_analysis["No_of_Bus_Operators"]
)

# Create Excel file with separate sheets
with pd.ExcelWriter("Top_10_Routes.xlsx", engine="openpyxl") as writer:

    for range_name, (low, high) in price_ranges.items():

        filtered = route_analysis[
            (route_analysis["Pricing"] >= low) &
            (route_analysis["Pricing"] < high)
        ]

        top_10 = filtered.sort_values(
            by="Opportunity_Score",
            ascending=False
        ).head(10)

        # Store each pricing range in separate sheet
        top_10.to_excel(
            writer,
            sheet_name=range_name,
            index=False
        )

print("Excel file 'Top_10_Routes.xlsx' created successfully.")
```

```python
# -------- UNDER-SERVED HIGH DEMAND ROUTES --------

# Create route analysis table
route_analysis = (
    df.groupby("Route")
    .agg({
        "Pricing": "mean",
        "Frequency": "first",
        "No_of_Bus_Operators": "first",
        "Distance": "mean"
    })
    .reset_index()
)

# Expansion Score
# High Frequency = Good
# High Pricing = Good
# More Operators = Bad

route_analysis["Expansion_Score"] = (
    route_analysis["Frequency"] * 0.5 +
    route_analysis["Pricing"] * 0.3 -
    route_analysis["No_of_Bus_Operators"] * 0.2
)

# Sort by best opportunity
best_routes = route_analysis.sort_values(
    by="Expansion_Score",
    ascending=False
)

# Top 20 under-served routes
top_20 = best_routes.head(20)

# Display result
print(top_20[[
    "Route",
    "Pricing",
    "Frequency",
    "No_of_Bus_Operators",
    "Distance",
    "Expansion_Score"
]])

# Save to CSV
top_20.to_csv(
    "UnderServed_HighDemand_Routes.csv",
    index=False
)

print("\nCSV file saved successfully.")
```

```python
import matplotlib.pyplot as plt

# Sort top 20 routes by Expansion Score
top_20 = top_20.sort_values(
    by="Expansion_Score",
    ascending=True
)

# Create figure
plt.figure(figsize=(14, 8))

# Horizontal bar chart
plt.barh(
    top_20["Route"],
    top_20["Expansion_Score"]
)

# Labels and title
plt.xlabel("Expansion Score")
plt.ylabel("Route")
plt.title("Top 20 Under-Served High Demand Routes")

# Better spacing
plt.tight_layout()

# Save visualization
plt.savefig(
    "UnderServed_HighDemand_Routes.png",
    dpi=300,
    bbox_inches="tight"
)

# Show plot
plt.show()

print("Visualization saved as 'UnderServed_HighDemand_Routes.png'")
```
